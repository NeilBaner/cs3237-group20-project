# Dumbell Exercise Classifier 

#### Description of Dataset
The dumbbell exercise dataset consits of a series of various dumbbell exercises along with their gyroscopic acceleration components and linear acceleration components as measured from the MPU9250. For each exercise there is certain threshold for that the components have to reach in order to detect muscle activation in the body. The following exercises are the eight different classes of this dataset.
1. Idle - Dumbbell not in motion
1. Bicep Curl - Used one label to distinguish motion from both hands as the orientation of the sensor remains the same and thus the vectors act in the same direction.
1. Shoulder-press - Used one label to distinguish motion from both shoulders as the orientation of the sensor remains the same and thus the vectors act in the same direction.
1. Left-shoulder-press - Used seperate label of left shoulder because the orientaion of the sensor will be unique to the left shoulder and the position of the vectors are acting in a specific direction. 
1. Right-shoulder-press - Used seperate label of right shoulder because the orientaion of the sensor will be unique to the right shoulder and the position of the vectors are acting in a specific direction.
1. Shoulder-front-raise - Used one label to distinguish motion from both hands as the orientation of the sensor remains the same and thus the vectors act in the same direction.
1. Forward-lunge - Used one label to distinguish motion from both legs as the orientation of the sensor remains the same and thus the vectors act in the same direction.
1. Dumbbell-squat - Used one label to distinguish motion from both hands as the orientation of the sensor remains the same and thus the vectors act in the same direction.
1. Left-tricep-extension - Used seperate label of left arm because the orientaion of the sensor will be unique to the left arm and the position of the vectors are acting in a specific direction.
1. Right-tricep-extension - Used seperate label of left arm because the orientaion of the sensor will be unique to the left arm and the position of the vectors are acting in a specific direction.
